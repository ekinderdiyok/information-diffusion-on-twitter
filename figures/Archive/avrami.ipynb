```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Define the Avrami equation function
def avrami_equation(t, k, n):
    return 1 - np.exp(-k * t ** n)

# Sample experimental data (time and fraction transformed)
time_data = np.array([0, 10, 20, 30, 40, 50, 60])
fraction_data = np.array([0, 0.2, 0.45, 0.65, 0.8, 0.9, 0.95])

# Fit the Avrami equation to the data
params, covariance = curve_fit(avrami_equation, time_data, fraction_data)

# Extract the fitted parameters
k_fit, n_fit = params

# Generate fitted curve for plotting
time_fit = np.linspace(0, 60, 100)
fraction_fit = avrami_equation(time_fit, k_fit, n_fit)

# Plot the experimental data and the fitted curve
plt.scatter(time_data, fraction_data, label='Experimental Data')
plt.plot(time_fit, fraction_fit, label='Fitted Curve')
plt.xlabel('Time (s)')
plt.ylabel('Fraction Transformed')
plt.legend()
plt.grid(True)
plt.show()

print(f"Fitted parameters: k = {k_fit}, n = {n_fit}")
```
